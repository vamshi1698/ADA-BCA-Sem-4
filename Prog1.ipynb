```python
import timeit
import matplotlib.pyplot as plt


def Add(array,n):
    for i in range(0,n):
        ele = int(input("Arr :"))
        array.append(ele)

def linear_search(array,key):
    for x in array:
        if x==key:
            return True
    return False

N = []
CPU = []
trail = int(input("Enter the no of trails:"))
for t in range(0,trail):
    array=[]
    print("---->trail no :",t+1)
    n=int(input("Enter the no of elements :"))
    Add(array,n)
    key = int(input("Enter key element:"))
    start=timeit.default_timer()
    s=linear_search(array,key)
    print("Element found = ",s)
    times = timeit.default_timer() - start
    N.append(n)
    CPU.append(round(float(times)*1000000,2))

print("N CPU")
for t in range(0,trail):
    print(N[t],CPU[t])
plt.plot(N,CPU)
plt.scatter(N,CPU,color="red",marker="*",s=50)
plt.xlabel("array size-N")
plt.ylabel("cpu processing time")
plt.title("linear search - VAMSI")
plt.show()
```
